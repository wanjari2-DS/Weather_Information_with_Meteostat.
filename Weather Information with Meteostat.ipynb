```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Monthly
```

```python
start = datetime(2020, 1, 1)
end = datetime(2023, 5, 20)

tpr = Point(21.146633, 79.088860)

data = Monthly(tpr, start, end)
data = data.fetch()
```

```python
# Plot line chart including average, minimum and maximum temperature
data.plot(y=['tavg', 'tmin', 'tmax'], 
          title='Temperature 2023 - Nagpur City')
plt.show()
```

```python
from meteostat import Stations
stations = Stations()
stations = stations.nearby(21.146633, 79.088860)
# Fetching 10 city/villages in nagpur
station = stations.fetch(10)
pd.DataFrame(station)
```
